# Data Inspection

This notebook covers the initial data inspection steps for the Heart Attack dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set display options
pd.set_option('display.max_columns', None)
sns.set_theme(style='whitegrid')
```

## Load Data

```python
# Load the raw dataset
df = pd.read_csv('../data/raw/heart_attack_dataset_raw.csv')
df.head()
```

## Dataset Shape

```python
print(f"Number of rows: {df.shape[0]}")
print(f"Number of columns: {df.shape[1]}")
```

## Basic Information

```python
# Basic info about data types and non-null counts
df.info()
```

## Summary Statistics

```python
# Summary statistics for numerical columns
df.describe()
```

```python
# Summary statistics for categorical columns
df.describe(include=['object'])
```

## Missing Values

```python
# Check for missing values
missing_data = df.isnull().sum()
missing_data = missing_data[missing_data > 0]
missing_data.sort_values(ascending=False)
```

## Data Duplicates

```python
# Check for duplicate rows
print(f"Number of duplicate rows: {df.duplicated().sum()}")
df[df.duplicated(keep=False)]
```

## Unique Values per Column

```python
# View the number of unique values for each column
df.nunique()
```
