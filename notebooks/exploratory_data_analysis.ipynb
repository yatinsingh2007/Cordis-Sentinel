# Exploratory Data Analysis (EDA)

Each feature is explored block-by-block with individual plots. Every numerical feature gets a histogram and a boxplot vs target. Every categorical feature gets a countplot and a grouped countplot vs target.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid', palette='muted')
plt.rcParams['figure.figsize'] = (8, 5)
```

```python
df = pd.read_csv('../data/interim/heart_attack_dataset_cleaned.csv')
print(f'Shape: {df.shape}')
df.head()
```

---
## 1. Target Variable — `heart_attack_risk`

```python
plt.figure(figsize=(6, 4))
ax = sns.countplot(data=df, x='heart_attack_risk', palette='Set2')
total = len(df)
for p in ax.patches:
    pct = f'{100 * p.get_height() / total:.1f}%'
    ax.annotate(pct, (p.get_x() + p.get_width() / 2, p.get_height()), ha='center', va='bottom')
plt.title('Distribution of Heart Attack Risk (Target)')
plt.xlabel('Heart Attack Risk (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

## 4. Correlation Heatmap

```python
num_cols_all = ['age', 'resting_bp', 'cholesterol', 'max_heart_rate', 'oldpeak',
                'num_major_vessels', 'bmi', 'stress_level', 'heart_attack_risk']

plt.figure(figsize=(12, 10))
corr_matrix = df[num_cols_all].corr()
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
sns.heatmap(corr_matrix, mask=mask, annot=True, fmt='.2f',
            cmap='coolwarm', vmin=-1, vmax=1, center=0, square=True)
plt.title('Correlation Heatmap — Numerical Features')
plt.show()
```
